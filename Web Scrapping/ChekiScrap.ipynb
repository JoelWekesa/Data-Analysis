```python
import requests as rqs
import pandas as pd 
import numpy as np
import seaborn as sns
import os
from bs4 import BeautifulSoup as BS
```

```python
cars = []
prices = []
links = []
```

```python
base = "https://www.cheki.co.ke/"
```

```python
for x in range(1, 10):
    source = rqs.get(f"https://www.cheki.co.ke/cars?page={x}").text
    soup = BS(source, "lxml")
    great = soup.find("div", class_="search-results__row")
    grand = great.find("div", class_="second search-results__main")
    father = grand.find("div", class_="tolSearchContainer")

    for child in father.find_all("li"):
        try:
            car = child.find("span", class_="ellipses").text 
            price = child.find("h2", class_="listing-price").text
            tail = child.find("div", class_="card-main").find("a")["href"]
            link = base + tail
        except AttributeError:
            car = np.nan
            price = np.nan
            tail = np.nan
            link = np.nan

        cars.append(car)
        prices.append(price)
        links.append(link)
```

```python
links = pd.Series(links)
cars = pd.Series(cars)
prices = pd.Series(prices) 
```

```python
df = pd.DataFrame() 
```

```python
df.insert(0, "Car", cars)
df.insert(1, "Price", prices)
df.insert(2, "Link", links) 
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df
```

```python
def removeline(item):
    item = item.replace("\n", "")
    return item 
```

```python
df["Car"] = df["Car"].apply(removeline) 
df["Price"] = df["Price"].apply(removeline) 
```
